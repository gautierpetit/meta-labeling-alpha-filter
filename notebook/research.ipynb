# Data selection:

S&P500 Point in time membership

Alleviates several biases: 
-  Look-Ahead Bias: Using today's S&P 500 list to backtest a strategy in the past assumes you knew which companies would survive and be in the index. Survivors tend to outperform — this introduces an optimistic bias.
- Survivorship Bias: Companies that went bankrupt or were removed from the index are excluded when using a static ticker list. This skews results upward by ignoring underperformers and delisted stocks.
- Selection Bias: If you only use stocks with complete long-term data, you're cherry-picking the data-rich survivors, again inflating performance.

Jegadeesh & Titman (1993) themselves show that momentum profits are stronger among past winners, but a look-ahead biased backtest tends to exaggerate this.

For meta-labeling and machine learning:
Models trained on biased historical data will learn patterns that didn’t exist at the time. This creates false confidence in predictive power and poor generalization.

Backtesting with a time accurate S&P500 universe provides a realistic look at how the US equity market truly was at a certain point in the past. It provides a more trustworthy depiction of the strategy performance for this time window by providing insights of how the strategy would have actually performed if implemeted at the time. Finally, from a meta learning quality point of view, the patterns learned by the algorithm are more robust and confers superior predictive power to unseen data.

# Base strategy

Classic Jegadeesh & Titman (1993) momentum strategy:

- Monthly signal generation: At the end of each month, compute each stocks 12-month past return (excluding the most recent month — often called a 1-month gap).

- Ranking: Rank all available stocks by their 12-month return.

- Top decile selection: Select top 10% (or top N) as the winners.

- Holding period: Hold the selected stocks for 3 months, equally weighted, with overlapping portfolios (i.e., rebalance monthly, but hold each signal for 3 months).

# Decisions on labeling definitions:

## Current:
+1 good trades: TP Hit for both long and short
0 timeout:
-1 bad trades: SL Hit for both long and short

🧠 What it teaches:
"Given that I took a long or short, tell me if this trade will succeed (TP), fail (SL), or expire."

✅ Pros:
Simple and symmetric

The model learns to detect good setups, regardless of direction

Useful if you want the model to act as a gatekeeper/meta-filter

❌ Cons:
Model has to learn two different patterns (rising long vs falling short) for one class (+1)

May reduce clarity when interpreting what a +1 prediction means

## Alternative Setup: Directionally-Split Success
+1 good long trade
0 timeout, failure, noise
-1 good short trade

🧠 What it teaches:
"Predict the type of opportunity this setup is: long, short, or nothing."

✅ Pros:
Very interpretable: a +1 prediction means go long, -1 means short

Can feed directly into signal generation — model predicts the position

Might improve learning since long/short setups are modeled separately

❌ Cons:
More difficult labeling logic

No distinction between failed and timed-out trades (unless you extend to 4 classes)

Harder to evaluate calibration or signal quality cleanly

# Decisions on signal usage:

# Current setup:
| Label | Meaning                            | Usage                           |
| ----- | ---------------------------------- | ------------------------------- |
| `+1`  | Trade hits TP → *successful trade* | Use `P(+1)` to keep signal      |
| `0`   | Timeout → *neutral trade*          | Ignored / neutral               |
| `-1`  | SL hit → *bad trade*               | Use `P(+1)` threshold to filter |


- Tier 1 – Use P(+1) only (Current Setup)
    - Already good, especially for:
    - Calibrated models
    - Binary meta-filtering
    - Easy interpretation

Keep this as your baseline.

- Tier 2 – Use gap logic P(+1) - max(P(-1), P(0)) ≥ gap
    - Improves robustness by guarding against:
    - Trades that are likely to time out
    - Ambiguous setups
    This is a smart middle ground.
    
You’re not ignoring the other classes, but still prioritizing clear confidence in success.

- Tier 3 – Weighted scaling using P(+1) * f(P(0)) decay
More advanced. Use when:
    - You want smoother sizing
    - Or already implemented class weighting in sizing.py
Use with caution:
    - Only if model calibration is very good
    - Can degrade performance if P(0) is noisy or uninformative

“Not all non-trades are equally unworthy — timeouts aren’t failures, and the model knows when it’s unsure.”
